In summary, these numbers are PyPI download totals for Numpy wheels from 30 days prior to 7 August 2019.

```python
import pandas as pd
```

```python
df = pd.read_csv('numpy_by_distro_version_id.csv')
df.head(20)
```

```python
by_version = (df.groupby(['distro_name', 'distro_version'])
.sum()
.sort_values('downloads', ascending=False)
             )
by_version.head(10)
```

```python
total_downloads = by_version['downloads'].sum()
total_downloads
```

```python
by_version['download_pct'] = by_version['downloads'] / total_downloads * 100
by_version.head(20)
```
